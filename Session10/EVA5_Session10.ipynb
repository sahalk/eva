```
# Importing the required module.

from eva.Session9.model import ResNet
from eva.Session9.loader import Loader
from eva.Session9.grad_cam import GradCam
```

```
# Mean and Std deviation of the dataset.
data_mean = (0.4914, 0.4822, 0.4465)
data_std_dev = (0.2470, 0.2435, 0.2616)

# Loading the dataset.
cifar = Loader(data_mean, data_std_dev)
```

```
# Displaying the images.
cifar.disp_image(cifar.trainloader)
```

```
# Summary of the model.
cifar.print_sum()
```

```
# Training the model.

learning_rate = 0.01
momentum = 0.9
epoch = 45

cifar.train(epoch, learning_rate, learning_rate)
```

```
# GradCam Module.
target_layers = ['layer1.0.bn2', 'layer1.1.bn2', 'layer2.0.bn2', 'layer2.1.bn2', 'layer3.0.bn2', 'layer3.1.bn2', 'layer4.0.bn2', 'layer4.1.bn2']
grad_cam = GradCam(cifar.model, cifar.testloader, target_layers=target_layers, device=cifar.device)
```

```
# GradCam for Car.

one_hot_encoded_class = 'car' 
required_class = 'car'
overlays = grad_cam.get_overlays(required_class=required_class, one_hot_encoded_class=one_hot_encoded_class)
print("Car")
grad_cam.display_heatmaps(overlays)
```

```
# GradCam for Truck.

one_hot_encoded_class = 'truck'
required_class = 'truck'
overlays = grad_cam.get_overlays(required_class=required_class, one_hot_encoded_class=one_hot_encoded_class)
print("Truck")
grad_cam.display_heatmaps(overlays)
```
